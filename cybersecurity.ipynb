```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import warnings
import plotly.express as px
```

```python
df=pd.read_csv("/content/Cyber_security.csv")
```

```python
import missingno as msno
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the data

# Set the figure size to make the visualization larger
plt.figure(figsize=(12, 8))

# Adjust the color scheme for better clarity
color = (0.2, 0.4, 0.8)  # Adjust color to your preference

# Plot the matrix with labels
msno.matrix(df, labels=True, color=color)

# Add a title to the plot for clarity
plt.title('Missing Values Matrix', fontsize=16)

# Display the plot
plt.show()
```

```python
msno.bar(df, color="dodgerblue", sort="ascending", labels=True, figsize=(40, 20))

```

```python
df.describe()
```

```python
corr=df.select_dtypes(include='number')
sns.heatmap(corr.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
data = df["Region"].value_counts()
plt.figure(figsize=(8,8))
plt.pie(data, labels=data.index,autopct='%.0f%%',shadow=True,wedgeprops={"width":0.3})
plt.title("Regions")
plt.show()

plt.figure(figsize=(8,8))
sns.countplot(x="Region", data=df, color='Maroon')
plt.title("Number of Countries of the Regions")
plt.show()

plt.figure(figsize=(8,8))
sns.histplot(df, x="Region", y="CEI",color="Maroon")
plt.title("Exposure to Cyber ​​Attacks by Region")
plt.show()

plt.figure(figsize=(8,8))
sns.histplot(df, x="Region", y="GCI",color="Maroon")
plt.title("Cyber ​​Security Commitment by Regions")
plt.show()

plt.figure(figsize=(8,8))
sns.histplot(df, x="Region", y="NCSI",color="Maroon")
plt.title("Readiness to Manage Cyber Incidents By Region")
plt.show()

plt.figure(figsize=(8,8))
sns.histplot(df, x="Region", y="DDL",color="Maroon")
plt.title("Average percentage of both CEI and GCI at their maximum value by Region")
plt.show()
```

```python
CEI = df.loc[:, ["Country", "CEI"]]
CEI["Size"] = CEI["CEI"].fillna(0.2)
fig = px.scatter_geo(CEI, locations='Country', locationmode='country names',
                     color='CEI', hover_name='Country', size="Size",
                     projection="natural earth", title='The value of the indicator "CEI" by country',
                     color_continuous_scale="sunset")

fig.update_layout(width=1000, height=800)
fig.show()
```

```python
GCI = df.loc[:, ["Country", "GCI"]]
GCI["Size"] = GCI["GCI"].fillna(0.2)

fig = px.scatter_geo(GCI, locations='Country', locationmode='country names',
                     color='GCI', hover_name='Country', size="Size",
                     projection="natural earth", title='The value of the indicator "GCI" by country',
                     color_continuous_scale="sunset")

fig.update_layout(width=1000, height=800)
fig.show()
```

```python
NCSI = df.loc[:, ["Country", "NCSI"]]
NCSI["Size"] = NCSI["NCSI"].fillna(0.2)

fig = px.scatter_geo(NCSI, locations='Country', locationmode='country names',
                     color='NCSI', hover_name='Country', size="Size",
                     projection="natural earth", title='The value of the indicator "NCSI" by country',
                     color_continuous_scale="sunset")

fig.update_layout(width=1000, height=800)
fig.show()
```

```python
DDL = df.loc[:, ["Country", "DDL"]]
DDL["Size"] = DDL["DDL"].fillna(0.2)

fig = px.scatter_geo(DDL, locations='Country', locationmode='country names',
                     color='DDL', hover_name='Country', size="Size",
                     projection="natural earth", title='The value of the indicator "DDL" by country',
                     color_continuous_scale="sunset")

fig.update_layout(width=1000, height=800)
fig.show()
```
